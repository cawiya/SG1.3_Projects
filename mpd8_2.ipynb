```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

# Highlight the key differences between KDE plots, box plots and violin plots
A violin plot is more informative than a plain box plot. While a box plot only shows summary statistics such as mean/median and interquartile ranges, the violin plot shows the full distribution of the data. 

A kernel density estimate (KDE) plot on the other hand is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

Relative to a histogram, KDE can produce a plot that is less cluttered and more interpretable, especially when drawing multiple distributions. But it has the potential to introduce distortions if the underlying distribution is bounded or not smooth. Like a histogram, the quality of the representation also depends on the selection of good smoothing parameters.



```python
sns.kdeplot(data=tips,shade=True)
plt.title('Size Feature of Tips Dataset')
plt.show()
```

This is a graphical representation of the tips dataset showing the distribution of total bill, tip and size of tips.

```python
#Distribution of total bill by gender for the tips dataset using the boxplots
sns.boxplot(data=tips, x='total_bill', y='sex')
plt.title('Distribution of Total bill by Gender')
plt.show()
```

The illustration above represents the distribution of total bill by gender and we can observe that the lowest bill for Male is around 7 while the lowest bill for female is around 3. Also, total bill for male gender is concentrated around 14 to 24 while for female, 12 to 21. Some outlier cases can also be observed on both genders.

```python
#Using boxplots generate the distribution of tips by days
sns.boxplot(data=tips, x='tip', y='day', palette='viridis')
plt.title('Distribution of tips by days')
plt.show()
```

The figure above shows a distribution of the tips received per day. It is observed that it can be as high as 10 on weekends.

```python
#Using boxplots generate the distribution of total bill by days considering smoker status of the customer
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker', palette= 'inferno')
plt.title('Total Bill by day considering smoker status')
plt.show()
```

```python
#Using violin plots generate the distribution of total bill by time given the genders
sns.violinplot(data=tips, y='time', x='total_bill', hue='sex', orient='horizontal', palette='plasma')
plt.title('Total bill by time based on Gender')
plt.show()
```

```python
#Generating a violin plot to show distribution of tips by day based on time
sns.violinplot(data=tips, y='day', x='tip', hue='time', orient='horizontal', palette='magma')
plt.title('Distribution of Tips by Day based on Time')
plt.show()
```

```python
#Plotting and customizing a heatmap for the tips Dataset
ax = sns.heatmap(tips.corr(), annot=True, fmt='.1f', cmap='PuRd', linewidths=8, cbar_kws={'shrink':.99})
plt.title('Correlation heatmap of tips Dataset \n'.title())
plt.show()  
```
