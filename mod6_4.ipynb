```python
#Perform an inner merge operation between customers and orders tables, using customer_id
df_customers_orders = df_customers.merge(df_orders, how='inner', on='customer_id')
df_customers_orders
```

In the table above, the merge function was applied between customer and orders dataset using a common value, customer ID. The output however started with the common value and the total number of rows are again limited to the number of rows which can provide a complete output across columns

```python
#Perform an outer merge between customers and products tables. using left_index,right_index
df_customers_products = pd.merge(df_customers,df_products,left_index=True,right_index=True,how ='outer')
df_customers_products
```

The customer and the product Dataframes had no similar columns however it was joined using the matching index of the DataFrames to which some values returned "Nan" due to the fact that there was no data in those fields under the customer dataset

```python
#Perform a left merge between orders and customers tables. picking customer_id for merge
df_orders_customers = pd.merge(df_orders,df_customers, how='left', on=['customer_id'])
df_orders_customers
```

The rows are kept in the left dataframe with missing values represented as NaN

```python
#Perform a right merge between product and customers tables

df_products_customers = pd.merge(df_products, df_customers,left_index=True,right_index=True,how = 'right')
df_products_customers
```
