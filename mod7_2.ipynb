```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Correlation between x and y using scatterplot and line plot in a subplot
#Four methods of creating subplots to visualize the data

#Method one
fig = plt.figure(figsize=(8, 6))
x = [0,14]
y = [0,10.84]
x1 = anscombe_df['x1']
y1 = anscombe_df['y1']
plt.subplot(2, 2, 1)
plt.title('(x1,y1)')
plt.xlabel('x1')
plt.ylabel('y1')
plt.scatter(x1, y1, color ="blue")
plt.plot(x,y, color = 'grey', linestyle = '-')

x = [0,14]
y = [0,9.26]
x2 = anscombe_df['x2']
y2 = anscombe_df['y2']
plt.subplot(2, 2, 2)
plt.title('(x2,y2)')
plt.xlabel('x2')
plt.ylabel('y2')
plt.scatter(x2, y2, color ="blue")
plt.plot(x,y, color = 'grey', linestyle = '-')

x = [0,14]
y = [0,12.74]
x3 = anscombe_df['x3']
y3 = anscombe_df['y3']
plt.subplot(2, 2, 3)
plt.title('(x3,y3)')
plt.xlabel('x3')
plt.ylabel('y3')
plt.scatter(x3, y3, color ="blue")
plt.plot(x,y, color = 'grey', linestyle = '-')

x = [0,19]
y = [0,12.5]
x4 = anscombe_df['x4']
y4 = anscombe_df['y4']
plt.subplot(2, 2, 4)
plt.title('(x4,y4)')
plt.xlabel('x4')
plt.ylabel('y4')
plt.scatter(x4, y4, color ="blue")
plt.plot(x,y, color = 'grey', linestyle = '-')

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
plt.show()
```

```python
#Method two
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

x = [0,14]
y = [0,10.84]
ax1.scatter(x1, y1, color ="green")
ax1.plot(x, y, color='grey')
ax1.set_title('(x1,y1)')
ax1.set_xlabel('x1')
ax1.set_ylabel('y1')

x = [0,14]
y = [0,9.26]
ax2.scatter(x2, y2, color ="green")
ax2.plot(x, y, color='grey')
ax2.set_title('(x2,y2)')
ax2.set_xlabel('x2')
ax2.set_ylabel('y2')


x = [0,14]
y = [0,12.74]
ax3.scatter(x3, y3, color ="green")
ax3.plot(x, y, color='grey')
ax3.set_title('(x3,y3)')
ax3.set_xlabel('x3')
ax3.set_ylabel('y3')


x = [0,19]
y = [0,12.5]
ax4.scatter(x4, y4, color ="green")
ax4.plot(x, y, color='grey')
ax4.set_title('(x4,y4)')
ax4.set_xlabel('x4')
ax4.set_ylabel('y4')
plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
```

```python
#Method three
fig = plt.figure(figsize=(8, 6))

x = [0,14]
y = [0,10.84]
ax5 = fig.add_subplot(221, title='(x1,y1)', xlabel='x1', ylabel='y1')
ax5.scatter(x1, y1, c ="red")
ax5.plot(x, y, color='grey')

x = [0,14]
y = [0,9.26]
ax6 = fig.add_subplot(222, title='(x2,y2)', xlabel='x2', ylabel='y2')
ax6.scatter(x2, y2, c ="red")
ax6.plot(x, y, color='grey')

x = [0,14]
y = [0,12.74]
ax7 = fig.add_subplot(223, title='(x3,y3)', xlabel='x3', ylabel='y3')
ax7.scatter(x3, y3, c ="red")
ax7.plot(x, y, color='grey')

x = [0,19]
y = [0,12.5]
ax8 = fig.add_subplot(224, title='(x4,y4)', xlabel='x4', ylabel='y4')
ax8.scatter(x4, y4, c ="red")
ax8.plot(x, y, color='grey')

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
```
