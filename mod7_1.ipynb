```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
sales_df = pd.read_csv('/content/drive/MyDrive/Stuternâ€™s Home Study Pack Submissions/company_sales_data.csv')
sales_df
```

```python
#Compare all the products sales data for every month using the multiple line plot
#Customize lines and include marker of choice

fig = plt.figure(figsize=(10,8))
x = sales_df['month_number']
y1 = sales_df['facecream']
plt.plot(x,y1, color = 'green', linestyle = '--', marker = 'o')

x = sales_df['month_number']
y2 = sales_df['toothpaste']
plt.plot(x,y2, color = 'blue', linestyle = '-', marker = '+')

x = sales_df['month_number']
y3 = sales_df['facewash']
plt.plot(x,y3, color = 'red', linestyle = ':')

x = sales_df['month_number']
y4 = sales_df['bathingsoap']
plt.plot(x,y4, color = 'grey', linestyle = 'solid', marker = 'o')

x = sales_df['month_number']
y5 = sales_df['shampoo']
plt.plot(x,y5, color = 'orange', linestyle = '-', marker = '+')

x = sales_df['month_number']
y6 = sales_df['moisturizer']
plt.plot(x,y6, color = 'black', linestyle = 'None', marker = '*')

#Put the legend at the top left corner of the axes
plt.legend(['facecream', 'toothpaste', 'facewash', 'bathingsoap', 'shampoo', 'moisturizer'],loc = 'upper left')
#Label the axes properly
plt.title("Cykim Skeen monthly Sales report")
plt.xlabel('Month number')
plt.ylabel("Sales per month")
#Display the results
plt.show()
```
