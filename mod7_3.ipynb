```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#import csv file
company_sales_df = pd.read_csv("/content/drive/MyDrive/company_sales_data.csv")
company_sales_df
```

```python
#Compare any two products sales data using a bar chart
x = np.arange(len(company_sales_df['month_number']))
month = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

ax = company_sales_df.plot.bar(x = 'month_number', y=['toothpaste', 'bathingsoap'], figsize = (15,10),  width = 0.7, color=['grey', 'pink'])
plt.legend(loc = 'upper left')
plt.title("Toothpaste and Bathing soap sales data")
plt.xlabel('Months')
plt.ylabel("Sales(Units)")
plt.xticks(x, month)
plt.savefig('compare_products_sales.png')
plt.show()
```

```python
#Assuming the average monthly sales of bathing soap is 10k, visualize the sales data of bathing soap of all months, showing how they differ from the average sale.
fig = plt.figure(figsize=(15, 5))
x = np.arange(len(company_sales_df['month_number']))
month = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
yerr = np.array(list(map(lambda x: 10000-x, company_sales_df['bathingsoap'])))
yerr = np.absolute(yerr)

y1 = company_sales_df['bathingsoap']
plt.bar(x,y1, yerr=yerr, capsize=10, label = 'bathingsoap',color='orange')

plt.title('Bathing Soap (Total to average monthly in units)')
plt.xlabel('Months')
plt.ylabel("Sales(units)")
plt.xticks(x, month)
plt.legend()
plt.savefig('Total-to-Average Monthly sales.png')
plt.show()
```

```python
#Calculate the total sales data for the year of  each product and represent it on a pie chart
fig = plt.gcf()
fig.set_size_inches(14,10) 
products_categories = ['Facecream', 'Facewash', 'Toothpaste', 'Bathingsoap','Shampoo','Moisturizer']
products_data = [company_sales_df['facecream'].sum(), company_sales_df['facewash'].sum(), company_sales_df['toothpaste'].sum(), 
                company_sales_df['bathingsoap'].sum(), company_sales_df['shampoo'].sum(), company_sales_df['moisturizer'].sum()]
plt.pie(products_data, labels=products_categories,autopct='%.d %%', colors= ['grey','pink','purple','orange','yellow', 'green'])
plt.title('Total product sales (quantity))')
plt.axis('equal')
plt.legend(loc = 'upper left')
plt.savefig('Total sales of each Product.png')
plt.show()
```
